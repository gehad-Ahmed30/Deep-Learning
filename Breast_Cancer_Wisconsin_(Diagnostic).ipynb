```python
import kagglehub
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.callbacks import EarlyStopping
```

```python
data=pd.read_csv(path+"/data.csv")
data.head()
```

```python
data.info()
```

```python
data.drop("Unnamed: 32",axis=1,inplace=True)
```

```python
data.drop("id",axis=1,inplace=True)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['diagnosis'].value_counts(normalize=True)
```

```python
sns.countplot(data['diagnosis'],palette="coolwarm")
plt.title("Distribution of Diagnosis (B vs M)")
plt.show()
```

```python
sns.pairplot(data, vars=["radius_mean", "perimeter_mean", "area_mean"], hue="diagnosis", palette="coolwarm")
plt.show()
```

```python
encoder=LabelEncoder()
data['diagnosis']=encoder.fit_transform(data['diagnosis'])
```

```python
x=data.drop('diagnosis',axis=1)
y=data['diagnosis']
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
X_standardized=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
(X_train,X_test,y_train,y_test)=train_test_split(X_standardized,y,test_size=0.2,random_state=42)
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(30,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=30)
```
